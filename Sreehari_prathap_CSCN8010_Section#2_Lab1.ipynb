The Objective of this Lab is to explore the fundamentals of machine learning by training a Univariate linear regression model. In Univariate regression model we are using one independent variable to predict the dependent variable.

### Goal

- median income
- population
- number of households

### Data Exploration
We are using the [California Housing Prices](https://www.kaggle.com/datasets/camnugent/california-housing-prices) Dataset from [Kaggle](https://www.kaggle.com/). The data contains information from the 1990 California census. It provides an accessible introductory dataset for learning about the basics of machine learning. The data pertains to the houses found in a given California district and some summary stats about them based on the 1990 census data. The columns are as follows:

1. **longitude** : A measure of how far west a house is; a higher value is farther west
2. **latitude** : A measure of how far north a house is; a higher value is farther north
3. **housingMedianAge** : Median age of a house within a block; a lower number is a newer building
4. **totalRooms** : Total number of rooms within a block
5. **totalBedrooms** : Total number of bedrooms within a block
6. **population** : Total number of people residing within a block
7. **households** : Total number of households, a group of people residing within a home unit, for a block
8. **medianIncome** : Median income for households within a block of houses (measured in tens of thousands of US Dollars)
9. **medianHouseValue** : Median house value for households within a block (measured in US Dollars)
10. **oceanProximity** : Location of the house w.r.t ocean/sea

```python
import pandas as pd

df = pd.read_csv('./housing.csv')
print(df.head())
```

```python
# Explore statistics using Pandas `describe` function
df[['median_house_value', 'median_income', 'population', 'households']].describe()
```

```python
# Create three Seaborn scatter plots of median house value vs. median income, population and number of households. 
# There should be one scatter plot per each independent variable. Add a title to each graph (using Seaborn). 
# Provide textual interpretation of the graphs, and insights specific to the linear regression you are going to conduct next.

# Import seaborn and matplotlib for plotting
import seaborn as sns
import matplotlib.pyplot as plt
```

### Scatter plot : median_house_value vs median_income

```python
# Scatter plot of median income vs median house value
sns.scatterplot(x='median_house_value', y='median_income', data=df)
# displaying the title
plt.title("median house value vs median income")
plt.show()
```

### Scatter plot : median_house_value vs population

```python
# Scatter plot of median income vs median house value
sns.scatterplot(x='median_house_value', y='population', data=df)
# displaying the title
plt.title("median house value vs population")
plt.show()
```

### Scatter plot : median_house_value vs households

```python
# Scatter plot of median income vs median house value
sns.scatterplot(x='median_house_value', y='households', data=df)
# displaying the title
plt.title("median house value vs no. of households")
plt.show()
```
